```python
#new feature
#Ask user for their name
name = input("What's your name?")
#Remove whitespace from string
name = name.strip()
#Capitalize's users name
name = name.capitalize()
#Say hello to user
print(f"Hello, {name}")
```

```python
#Ask user for their name
name = input("What's your name?")
#Remove whitespace from string
name = name.strip()
#Capitalize's users name
name = name.title()
#Say hello to user
print(f"Hello, {name}")
```

```python
#Ask user for their name
name = input("What's your name?")
#Remove whitespace from string and Capitalize's users name
name = name.strip().title()
#Say hello to user
print(f"Hello, {name}")
```

```python
#Ask user for their name
name = input("What's your name?").strip().title()
#Say hello to user
print(f"Hello, {name}")
```

```python
#Ask user for their name
name = input("What's your name?").strip().title()
# Split user's name into first name and last name
name = name.split()
#Say hello to user
print(f"Hello, {name}")
```
